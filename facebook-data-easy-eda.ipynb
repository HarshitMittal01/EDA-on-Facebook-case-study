# 6.Data preprocessing

```python
fbdata.isnull().sum()
```

We have 2 columns with missing values

### Dropping the missing values in tenure 

```python
fbdata = fbdata[~fbdata.tenure.isnull()].copy()

fbdata.isnull().sum()
```

### Gender ratio between male and female users

```python
x = fbdata['gender']=='male'
males = x.value_counts()[True]
non_male = x.value_counts()[False]

y = fbdata['gender']=='female'
females = y.value_counts()[True]

nan_count = non_male - females
total = males + non_male
print('males :',males)
print('females:',females)
print('nan count:', nan_count)
print('total:',total)
print('ratio - male:female :' ,males/females)
```

### Filling the nan values

```python
fbdata.gender.fillna(value="not revealed",inplace=True)
```

```python
fbdata.isnull().sum()
```

```python
x = fbdata['gender']=='male'
males = x.value_counts()[True]

y = fbdata['gender']=='female'
females = y.value_counts()[True]

z = fbdata['gender']=='not revealed'
not_revealed = z.value_counts()[True]

total = males + females + not_revealed

print('males :',males)
print('females:',females)
print('not_revealed:', not_revealed)
print('total:',total)
```

## Gender distribution through pie chart

```python
plt.pie(fbdata.gender.value_counts(),labels=['male','female','not revealed'],explode=(0,0,0),autopct='%1.1f%%')
plt.show()
```
